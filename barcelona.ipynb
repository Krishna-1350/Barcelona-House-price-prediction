Considering pandemic

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.read_csv('Barcelona_rent_price.csv')

df['Time'] = df['Year'] + df['Trimester']*0.25
ar = df[df['Average _rent']=='average rent (euro/month)']
data = ar.drop(['Year', 'Trimester', 'Neighbourhood', 'Average _rent'], axis=1)
data['Time'] = (data['Time']-2014.25)/8.25

models = []
dist = list(data['District'].unique())

for i in range(10):
  x = data[data['District']==dist[i]]
  X = x['Time'].to_numpy().reshape(-1,1)
  y = x['Price'].to_numpy().reshape(-1,1)
  regr = linear_model.LinearRegression()
  regr.fit(X, y)
  models.append(regr)

test = np.linspace(0,1,34)
pred = []
for i in range(10):
  pred.append(models[i].predict(test.reshape(-1,1)))

def prediction(district, year, trimester):
  return models[dist.index(district)].predict([[(year+0.25*trimester - 2014.25)/8.25]])[0][0]

plt.figure(figsize=(28,8))
y = ar['District'].unique()
for i in range(10):
  plt.subplot(2,5,i+1)
  x = ar[ar['District']==y[i]]
  plt.plot(test,pred[i])
  plt.plot((x['Time'].unique()-2014.25)/8.25, x.groupby(['Time'])['Price'].mean())
  plt.title(y[i])
plt.plot()
```

Neglecting pandemic

```python
ar = df[df['Average _rent']=='average rent (euro/month)']
data = ar.drop(['Year', 'Trimester', 'Neighbourhood', 'Average _rent'], axis=1)
data['Time'] = (data['Time']-2014.25)/6
data = data[data['Time']<=1.0]
models2 = []
dist = list(data['District'].unique())

for i in range(10):
  x = data[data['District']==dist[i]]
  X = x['Time'].to_numpy().reshape(-1,1)
  y = x['Price'].to_numpy().reshape(-1,1)
  regr2 = linear_model.LinearRegression()
  regr2.fit(X, y)
  models2.append(regr2)

test = np.linspace(0,1,25)
pred2 = []
for i in range(10):
  pred2.append(models2[i].predict(test.reshape(-1,1)))

def prediction2(district, year, trimester):
  return models2[dist.index(district)].predict([[(year+0.25*trimester - 2014.25)/6]])[0][0]

plt.figure(figsize=(28,8))
y = ar['District'].unique()
for i in range(10):
  plt.subplot(2,5,i+1)
  x = data[data['District']==y[i]]
  plt.plot(test,pred2[i])
  plt.plot(x['Time'].unique(), x.groupby(['Time'])['Price'].mean())
  plt.title(y[i])
plt.plot()
```
